```python
import qiskit_nature
from qiskit.algorithms.minimum_eigensolvers import NumPyMinimumEigensolver, VQE
from qiskit_nature.second_q.transformers import FreezeCoreTransformer
from qiskit_nature.second_q.formats.molecule_info import MoleculeInfo
from qiskit_nature.second_q.mappers import ParityMapper
from qiskit_nature.second_q.circuit.library import UCCSD, HartreeFock
from qiskit.algorithms.optimizers import SLSQP
from qiskit_nature.second_q.drivers import PySCFDriver
import matplotlib.pyplot as plt
from qiskit_aer.primitives import Estimator
import numpy as np

# It's good practice to disable the PauliSumOp setting for modern Qiskit Nature
qiskit_nature.settings.use_pauli_sum_op = False

def get_qubit_op(scale_factor):
    """
    Generates the qubit operator and other necessary information for the CH4 molecule
    at a given C-H bond length scale factor.
    """
    # Define methane molecule with scaling factor for C-H bond length
    # Standard C-H bond length is around 1.09 Angstrom.
    base_bond_length = 1.09
    bond_length = base_bond_length * scale_factor

    # Calculate H atom positions for a tetrahedron centered at the origin
    d = bond_length / np.sqrt(3)
    
    molecule = MoleculeInfo(
        symbols=["C", "H", "H", "H", "H"],
        coords=(
            [0.0, 0.0, 0.0],      # Carbon at the origin
            [d, d, d],            # Hydrogen 1
            [d, -d, -d],          # Hydrogen 2
            [-d, d, -d],          # Hydrogen 3
            [-d, -d, d]           # Hydrogen 4
        ),
        multiplicity=1,  # Singlet ground state
        charge=0,
    )

    driver = PySCFDriver.from_molecule(molecule)
    properties = driver.run()

    # Freeze the 1s core orbital AND remove the 3 highest virtual orbitals
    # This reduces the problem size for a much faster calculation, with a
    # minor trade-off in accuracy.
    problem = FreezeCoreTransformer(
        freeze_core=True, remove_orbitals=[6, 7, 8]
    ).transform(properties)

    num_particles = problem.num_particles
    num_spatial_orbitals = problem.num_spatial_orbitals

    mapper = ParityMapper(num_particles=num_particles)
    qubit_op = mapper.map(problem.second_q_ops()[0])

    return qubit_op, num_particles, num_spatial_orbitals, problem, mapper

def exact_solver(qubit_op, problem):
    """Computes the exact ground state energy using a classical eigensolver."""
    sol = NumPyMinimumEigensolver().compute_minimum_eigenvalue(qubit_op)
    result = problem.interpret(sol)
    return result

# --- Main Calculation ---

# Define a range of scaling factors for the C-H bond lengths
scale_factors = np.linspace(0.6, 2.0, 15)
exact_energies = []
vqe_energies = []
optimizer = SLSQP(maxiter=15)
noiseless_estimator = Estimator(approximation=True)

print("⚛️  Starting calculations for CH₄ molecule (symmetric stretch)...")
for scale in scale_factors:
    print(f"\nProcessing scale factor: {scale:.2f}")
    try:
        (qubit_op, num_particles, num_spatial_orbitals, problem, mapper) = get_qubit_op(scale)

        # --- Exact solution (for comparison within the reduced active space) ---
        result = exact_solver(qubit_op, problem)
        exact_energy = result.total_energies[0].real
        exact_energies.append(exact_energy)

        # --- VQE solution ---
        init_state = HartreeFock(num_spatial_orbitals, num_particles, mapper)
        var_form = UCCSD(
            num_spatial_orbitals, num_particles, mapper, initial_state=init_state
        )

        vqe = VQE(
            noiseless_estimator,
            var_form,
            optimizer,
            initial_point=[0] * var_form.num_parameters,
        )
        vqe_calc = vqe.compute_minimum_eigenvalue(qubit_op)
        vqe_result = problem.interpret(vqe_calc).total_energies[0].real
        vqe_energies.append(vqe_result)

        print(f"  C-H Bond Length: {scale * 1.09:.3f} Å")
        print(f"  VQE Energy:      {vqe_result:.5f} Hartree")
        print(f"  Exact Energy:    {exact_energy:.5f} Hartree")

    except Exception as e:
        print(f"Error at scale factor {scale}: {e}")
        # Add None or NaN values to maintain array structure if an error occurs
        exact_energies.append(np.nan)
        vqe_energies.append(np.nan)

print("\n✅ All energies have been calculated.")

# --- Plotting the Results ---

# Convert scale factors to actual bond lengths for the x-axis
bond_lengths = scale_factors * 1.09

plt.figure(figsize=(10, 6))
plt.plot(bond_lengths, exact_energies, "o-", color='limegreen', linewidth=2, label="Exact Energy (FCI in Active Space)")
plt.plot(bond_lengths, vqe_energies, "x--", color='red', linewidth=2, label="VQE Energy (UCCSD)")
plt.xlabel("C-H Bond Length (Angstrom)")
plt.ylabel("Energy (Hartree)")
plt.title("CH₄ Potential Energy Surface (Symmetric Stretch)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import time
from datetime import datetime

# --- Classical Chemistry Imports ---
from pyscf import gto, scf, fci

# --- Qiskit Legacy Imports (for qiskit-nature 0.6.0) ---
from qiskit_nature.drivers.second_quantization import PySCFDriver
from qiskit_nature.problems.second_quantization.electronic import ElectronicStructureProblem
from qiskit_nature.properties.second_quantization.electronic import ParticleNumber
from qiskit_nature.converters.second_quantization import QubitConverter
from qiskit_nature.mappers.second_quantization import ParityMapper
from qiskit_nature.transformers.second_quantization.electronic import FreezeCoreTransformer
from qiskit_nature.circuit.library import HartreeFock, UCCSD

# Qiskit Algorithm and Backend imports
from qiskit.algorithms import VQE
from qiskit.algorithms.optimizers import L_BFGS_B
from qiskit_aer import Aer
from qiskit.utils import QuantumInstance

# ----- Configuration -----
BASIS_SET = "sto-3g"

ATOM_GEOMETRY_STR = """C 0.00000000 0.00000000 0.00000000
H 0.62863435 0.62863435 0.62863435
H -0.62863435 -0.62863435 0.62863435
H -0.62863435 0.62863435 -0.62863435
H 0.62863435 -0.62863435 -0.62863435"""

ATOM_GEOMETRY_LIST = [
    ['C', [0.00000000, 0.00000000, 0.00000000]],
    ['H', [0.62863435, 0.62863435, 0.62863435]],
    ['H', [-0.62863435, -0.62863435, 0.62863435]],
    ['H', [-0.62863435, 0.62863435, -0.62863435]],
    ['H', [0.62863435, -0.62863435, -0.62863435]],
]

# ----- Main Calculation Function -----
def main():
    """
    Calculates the ground state energy of CH4 using FCI and VQE.
    """
    print(f"Methane (CH₄) Ground State Calculation (for Qiskit 0.42.1)")
    print(f"Started at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")

    results = {}

    # --- 1. Classical FCI (Exact) Calculation with PySCF ---
    print("--- Running Classical FCI Calculation ---")
    start_time = time.time()
    
    mol = gto.Mole(atom=ATOM_GEOMETRY_LIST, basis=BASIS_SET, verbose=0)
    mol.build()
    mf_rhf = scf.RHF(mol).run()
    fci_solver = fci.FCI(mol, mf_rhf.mo_coeff)
    results['FCI'] = fci_solver.kernel()[0]
    
    print(f"FCI (Exact) Energy: {results['FCI']:.8f} Hartree")
    print(f"Classical calculation finished in {time.time() - start_time:.2f}s\n")

    # --- 2. Qiskit Nature VQE Calculation (Manual Workflow) ---
    print("--- Setting up Qiskit Nature Problem ---")

    driver = PySCFDriver(atom=ATOM_GEOMETRY_STR, basis=BASIS_SET)
    properties = driver.run()
    
    transformer = FreezeCoreTransformer(freeze_core=True)
    properties_transformed = transformer.transform(properties)

    particle_number = properties_transformed.get_property(ParticleNumber)
    num_particles = (particle_number.num_alpha, particle_number.num_beta)
    num_spin_orbitals = particle_number.num_spin_orbitals
    num_spatial_orbitals = num_spin_orbitals // 2
    
    # Step A: Manually get the second-quantized operators
    second_q_ops = properties_transformed.second_q_ops()
    # **THE FINAL FIX**: Access the operator by its dictionary key 'ElectronicEnergy'
    main_second_q_op = second_q_ops['ElectronicEnergy'] 

    # Step B: Manually convert to a qubit operator
    mapper = ParityMapper()
    converter = QubitConverter(mapper=mapper, two_qubit_reduction=True)
    qubit_op = converter.convert(main_second_q_op, num_particles=num_particles)

    # Step C: Build the ansatz
    initial_state = HartreeFock(num_spatial_orbitals, num_particles, converter)
    ansatz = UCCSD(
        qubit_converter=converter,
        num_particles=num_particles,
        num_spin_orbitals=num_spin_orbitals,
        initial_state=initial_state
    )

    print(f"Problem properties determined. Qubit operator has {qubit_op.num_qubits} qubits.")

    # Step D: Run the VQE algorithm directly
    optimizer = L_BFGS_B(maxiter=1000)
    backend = Aer.get_backend('statevector_simulator')
    quantum_instance = QuantumInstance(backend)
    
    vqe = VQE(ansatz, optimizer, quantum_instance=quantum_instance)

    print("\n--- Running VQE Calculation ---")
    start_time = time.time()
    vqe_result = vqe.compute_minimum_eigenvalue(operator=qubit_op)

    # Step E: Interpret the raw VQE result to get the energy
    problem_for_interpretation = ElectronicStructureProblem(driver, [transformer])
    final_result = problem_for_interpretation.interpret(vqe_result)
    results['VQE'] = final_result.total_energies[0].real

    print(f"VQE Energy: {results['VQE']:.8f} Hartree")
    print(f"VQE calculation finished in {time.time() - start_time:.2f}s\n")

    # --- 3. Final Summary ---
    print("\n" + "="*50)
    print("             Final Energy Summary")
    print("="*50)
    print(f"  VQE Energy:         {results['VQE']:.8f} Hartree")
    print(f"  FCI (Exact) Energy: {results['FCI']:.8f} Hartree")
    print("-" * 50)
    print(f"  Error (VQE vs FCI): {abs(results['VQE'] - results['FCI']):.8f} Hartree")
    print("="*50)

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
"""
Quantum Simulation of CH4 (Methane) Molecule
Using Qiskit Nature and PySCF for quantum chemistry calculations
"""

import numpy as np
from qiskit import QuantumCircuit, Aer, execute
from qiskit.circuit.library import RealAmplitudes, TwoLocal
from qiskit.algorithms import VQE, NumPyEigensolver
from qiskit.algorithms.optimizers import SLSQP, COBYLA, SPSA
from qiskit.primitives import Estimator
from qiskit.opflow import PauliSumOp, StateFn, I, X, Y, Z
from qiskit_nature.second_q.drivers import PySCFDriver
from qiskit_nature.second_q.mappers import JordanWignerMapper, ParityMapper, BravyiKitaevMapper
from qiskit_nature.second_q.circuit.library import HartreeFock, UCCSD
from qiskit_nature.second_q.problems import ElectronicStructureProblem
from qiskit_nature.second_q.transformers import ActiveSpaceTransformer
import matplotlib.pyplot as plt
from pyscf import gto, scf, mcscf, fci
import warnings
warnings.filterwarnings('ignore')

class CH4QuantumSimulation:
    """
    Complete quantum simulation framework for CH4 molecule
    """
    
    def __init__(self, basis='sto-3g', charge=0, spin=0):
        """
        Initialize the CH4 quantum simulation
        
        Args:
            basis: Basis set for quantum chemistry calculation
            charge: Molecular charge
            spin: Spin multiplicity (2S)
        """
        self.basis = basis
        self.charge = charge
        self.spin = spin
        self.geometry = self._get_ch4_geometry()
        self.driver = None
        self.problem = None
        self.hamiltonian = None
        self.mapper = None
        
    def _get_ch4_geometry(self):
        """
        Returns optimized CH4 geometry in Angstroms
        Tetrahedral structure with C-H bond length ~1.09 Å
        """
        # Tetrahedral geometry for CH4
        bond_length = 1.09  # Angstroms
        angle = np.arccos(-1/3)  # Tetrahedral angle
        
        geometry = [
            ['C', [0.0000, 0.0000, 0.0000]],
            ['H', [bond_length, 0.0000, 0.0000]],
            ['H', [-bond_length/3, bond_length*np.sqrt(8/9), 0.0000]],
            ['H', [-bond_length/3, -bond_length*np.sqrt(2/9), bond_length*np.sqrt(2/3)]],
            ['H', [-bond_length/3, -bond_length*np.sqrt(2/9), -bond_length*np.sqrt(2/3)]]
        ]
        return geometry
    
    def setup_molecular_problem(self):
        """
        Set up the molecular electronic structure problem
        """
        print("Setting up CH4 molecular problem...")
        print(f"Geometry: {self.geometry}")
        print(f"Basis set: {self.basis}")
        print(f"Charge: {self.charge}, Spin: {self.spin}")
        
        # Create PySCF driver
        self.driver = PySCFDriver(
            atom=self.geometry,
            basis=self.basis,
            charge=self.charge,
            spin=self.spin,
            unit='Angstrom'
        )
        
        # Run the driver to get molecular data
        self.problem = self.driver.run()
        
        # Get molecular information
        print("\nMolecular Information:")
        print(f"Number of electrons: {self.problem.num_particles}")
        print(f"Number of molecular orbitals: {self.problem.num_spatial_orbitals}")
        
        return self.problem
    
    def classical_solution(self):
        """
        Compute classical solution using PySCF
        """
        print("\n" + "="*50)
        print("Classical Solution (PySCF)")
        print("="*50)
        
        # Build PySCF molecule
        mol = gto.Mole()
        mol.atom = self.geometry
        mol.basis = self.basis
        mol.charge = self.charge
        mol.spin = self.spin
        mol.unit = 'Angstrom'
        mol.build()
        
        # Hartree-Fock calculation
        mf = scf.RHF(mol)
        hf_energy = mf.kernel()
        print(f"Hartree-Fock Energy: {hf_energy:.8f} Ha")
        
        # Configuration Interaction (if small enough)
        if mol.nao < 10:
            cisolver = fci.FCI(mf)
            fci_energy = cisolver.kernel()[0]
            print(f"Full CI Energy: {fci_energy:.8f} Ha")
        else:
            print("System too large for Full CI calculation")
            fci_energy = None
            
        return hf_energy, fci_energy
    
    def map_to_qubit_hamiltonian(self, mapper_type='jordan_wigner', active_space=None):
        """
        Map the fermionic Hamiltonian to qubit Hamiltonian
        
        Args:
            mapper_type: Type of fermion-to-qubit mapping
            active_space: Tuple of (num_electrons, num_orbitals) for active space
        """
        print("\n" + "="*50)
        print("Mapping to Qubit Hamiltonian")
        print("="*50)
        
        # Select mapper
        if mapper_type == 'jordan_wigner':
            self.mapper = JordanWignerMapper()
        elif mapper_type == 'parity':
            self.mapper = ParityMapper()
        elif mapper_type == 'bravyi_kitaev':
            self.mapper = BravyiKitaevMapper()
        else:
            raise ValueError(f"Unknown mapper type: {mapper_type}")
        
        print(f"Using {mapper_type} mapping")
        
        # Apply active space transformation if specified
        if active_space:
            print(f"Applying active space: {active_space[0]} electrons in {active_space[1]} orbitals")
            transformer = ActiveSpaceTransformer(
                num_electrons=active_space[0],
                num_molecular_orbitals=active_space[1]
            )
            self.problem = transformer.transform(self.problem)
        
        # Get the second quantized Hamiltonian
        second_q_op = self.problem.hamiltonian.second_q_op()
        
        # Map to qubit operator
        self.hamiltonian = self.mapper.map(second_q_op)
        
        print(f"Number of qubits required: {self.hamiltonian.num_qubits}")
        print(f"Number of Pauli terms: {len(self.hamiltonian)}")
        
        return self.hamiltonian
    
    def create_ansatz(self, ansatz_type='uccsd'):
        """
        Create quantum circuit ansatz
        
        Args:
            ansatz_type: Type of ansatz ('uccsd', 'hf', 'real_amplitudes')
        """
        print("\n" + "="*50)
        print("Creating Quantum Circuit Ansatz")
        print("="*50)
        
        num_particles = self.problem.num_particles
        num_spatial_orbitals = self.problem.num_spatial_orbitals
        
        # Create initial state (Hartree-Fock)
        init_state = HartreeFock(
            num_spatial_orbitals=num_spatial_orbitals,
            num_particles=num_particles,
            qubit_mapper=self.mapper
        )
        
        if ansatz_type == 'uccsd':
            # UCCSD ansatz
            ansatz = UCCSD(
                num_spatial_orbitals=num_spatial_orbitals,
                num_particles=num_particles,
                qubit_mapper=self.mapper,
                initial_state=init_state
            )
            print(f"UCCSD ansatz created with {ansatz.num_parameters} parameters")
            
        elif ansatz_type == 'hf':
            # Just Hartree-Fock state
            ansatz = init_state
            print("Using Hartree-Fock initial state only")
            
        elif ansatz_type == 'real_amplitudes':
            # Hardware-efficient ansatz
            ansatz = RealAmplitudes(
                num_qubits=self.hamiltonian.num_qubits,
                reps=3,
                entanglement='linear',
                initial_state=init_state
            )
            print(f"RealAmplitudes ansatz with {ansatz.num_parameters} parameters")
            
        else:
            raise ValueError(f"Unknown ansatz type: {ansatz_type}")
        
        return ansatz
    
    def run_vqe(self, ansatz=None, optimizer=None, initial_point=None):
        """
        Run Variational Quantum Eigensolver (VQE)
        
        Args:
            ansatz: Quantum circuit ansatz
            optimizer: Classical optimizer
            initial_point: Initial parameter values
        """
        print("\n" + "="*50)
        print("Running VQE Simulation")
        print("="*50)
        
        if ansatz is None:
            ansatz = self.create_ansatz('uccsd')
        
        if optimizer is None:
            optimizer = SLSQP(maxiter=100)
            print("Using SLSQP optimizer")
        
        # Create VQE instance
        estimator = Estimator()
        vqe = VQE(
            estimator=estimator,
            ansatz=ansatz,
            optimizer=optimizer,
            initial_point=initial_point
        )
        
        # Run VQE
        print("Starting VQE optimization...")
        result = vqe.compute_minimum_eigenvalue(self.hamiltonian)
        
        print("\nVQE Results:")
        print(f"Ground state energy: {result.eigenvalue:.8f} Ha")
        print(f"Optimization converged: {result.optimizer_result.fun is not None}")
        print(f"Number of function evaluations: {result.optimizer_result.nfev}")
        
        return result
    
    def analyze_results(self, vqe_result, classical_energies):
        """
        Analyze and compare quantum and classical results
        """
        print("\n" + "="*50)
        print("Results Analysis")
        print("="*50)
        
        vqe_energy = vqe_result.eigenvalue
        hf_energy, fci_energy = classical_energies
        
        print(f"VQE Energy: {vqe_energy:.8f} Ha")
        print(f"Hartree-Fock Energy: {hf_energy:.8f} Ha")
        
        if fci_energy is not None:
            print(f"Full CI Energy: {fci_energy:.8f} Ha")
            print(f"VQE Error vs FCI: {abs(vqe_energy - fci_energy):.8f} Ha")
        
        print(f"Correlation Energy (VQE): {vqe_energy - hf_energy:.8f} Ha")
        
        return {
            'vqe_energy': vqe_energy,
            'hf_energy': hf_energy,
            'fci_energy': fci_energy,
            'correlation_energy': vqe_energy - hf_energy
        }
    
    def visualize_circuit(self, ansatz):
        """
        Visualize the quantum circuit
        """
        print("\n" + "="*50)
        print("Circuit Visualization")
        print("="*50)
        
        # Decompose the circuit for visualization
        circuit = ansatz.decompose()
        print(f"Circuit depth: {circuit.depth()}")
        print(f"Number of gates: {circuit.size()}")
        print(f"Number of qubits: {circuit.num_qubits}")
        
        # Draw circuit (text format for compatibility)
        print("\nCircuit diagram (first 5 layers):")
        print(circuit.draw(output='text', fold=80, initial_state=True, 
                          vertical_compression='high', idle_wires=False))
        
        return circuit

def main():
    """
    Main execution function for CH4 quantum simulation
    """
    print("="*60)
    print("QUANTUM SIMULATION OF CH4 (METHANE) MOLECULE")
    print("="*60)
    
    # Initialize simulation with minimal basis set for feasibility
    # For production, use larger basis sets like '6-31g' or 'cc-pvdz'
    sim = CH4QuantumSimulation(basis='sto-3g', charge=0, spin=0)
    
    # Step 1: Set up molecular problem
    problem = sim.setup_molecular_problem()
    
    # Step 2: Get classical solution for comparison
    classical_energies = sim.classical_solution()
    
    # Step 3: Map to qubit Hamiltonian
    # Use active space to reduce computational cost
    # CH4 has 10 electrons, we'll use active space of (4, 4) for demonstration
    hamiltonian = sim.map_to_qubit_hamiltonian(
        mapper_type='jordan_wigner',
        active_space=(4, 4)  # 4 electrons in 4 orbitals
    )
    
    # Step 4: Create ansatz
    ansatz = sim.create_ansatz(ansatz_type='uccsd')
    
    # Step 5: Visualize circuit
    circuit = sim.visualize_circuit(ansatz)
    
    # Step 6: Run VQE
    vqe_result = sim.run_vqe(
        ansatz=ansatz,
        optimizer=SLSQP(maxiter=50)  # Reduced iterations for demo
    )
    
    # Step 7: Analyze results
    results = sim.analyze_results(vqe_result, classical_energies)
    
    print("\n" + "="*60)
    print("SIMULATION COMPLETE")
    print("="*60)
    
    return sim, results

if __name__ == "__main__":
    # Run the simulation
    simulation, results = main()
    
    # Additional analysis or custom experiments can be added here
    print("\nYou can now use the 'simulation' object for further experiments:")
    print("- Try different basis sets: simulation.basis = '6-31g'")
    print("- Try different mappers: 'parity', 'bravyi_kitaev'")
    print("- Try different ansätze: 'real_amplitudes', 'hf'")
    print("- Modify active space for accuracy vs. performance trade-off")
```

```python
import numpy as np
from pyscf import gto, scf

# Qiskit imports
from qiskit import Aer
from qiskit.algorithms import VQE, NumPyMinimumEigensolver
from qiskit.algorithms.optimizers import COBYLA
from qiskit.circuit.library import EfficientSU2
from qiskit.utils import QuantumInstance

# Qiskit Nature imports
from qiskit_nature.drivers import Molecule
from qiskit_nature.drivers.second_quantization import PySCFDriver
from qiskit_nature.problems.second_quantization import ElectronicStructureProblem
from qiskit_nature.converters.second_quantization import QubitConverter
from qiskit_nature.mappers.second_quantization import JordanWignerMapper
from qiskit_nature.algorithms import GroundStateEigensolver
from qiskit_nature.transformers import FreezeCoreTransformer

# Define the methane (CH4) geometry in Angstroms
methane = Molecule(
    geometry=[
        ['C', [0.0, 0.0, 0.0]],
        ['H', [0.629118, 0.629118, 0.629118]],
        ['H', [-0.629118, -0.629118, 0.629118]],
        ['H', [0.629118, -0.629118, -0.629118]],
        ['H', [-0.629118, 0.629118, -0.629118]]
    ],
    charge=0,
    multiplicity=1
)

print("Molecule geometry defined:")
print(methane)

# Create the driver with PySCF
driver = PySCFDriver.from_molecule(methane, basis='sto-3g')

# Apply the frozen core approximation - freeze 1s orbitals of carbon
freeze_core = FreezeCoreTransformer()

# Construct the electronic structure problem with frozen core
problem = ElectronicStructureProblem(driver, transformers=[freeze_core])
print("\nPreparing problem with frozen core approximation...")

# Create a qubit converter using Jordan-Wigner mapping
qubit_converter = QubitConverter(JordanWignerMapper())

# Get the number of qubits needed
second_q_ops = problem.second_q_ops()
main_op = second_q_ops[0]
qubit_op = qubit_converter.convert(main_op)
num_qubits = qubit_op.num_qubits
print(f"Number of qubits required (with frozen core): {num_qubits}")

# Run classical exact solver first for reference
classical_solver = NumPyMinimumEigensolver()
classical_groundstate = GroundStateEigensolver(qubit_converter, classical_solver)
classical_result = classical_groundstate.solve(problem)
print(f"Reference energy (classical): {classical_result.total_energies[0].real:.8f} Hartrees")

# Set up quantum solver with VQE
quantum_instance = QuantumInstance(
    Aer.get_backend('statevector_simulator'),
    shots=1024,
    seed_simulator=42,
    seed_transpiler=42
)

# Use efficient SU2 ansatz for better energy convergence
ansatz = EfficientSU2(num_qubits, reps=2, entanglement='linear')
optimizer = COBYLA(maxiter=100)

# Set up VQE
vqe = VQE(ansatz, optimizer, quantum_instance=quantum_instance)
vqe_solver = GroundStateEigensolver(qubit_converter, vqe)

# Solve using VQE
print("\nComputing CH4 ground state energy using VQE with frozen core...")
print("This may take a few minutes...")
vqe_result = vqe_solver.solve(problem)

# Display results
print("\n=== RESULTS ===")
print(f"VQE Energy: {vqe_result.total_energies[0].real:.8f} Hartrees")
print(f"Reference Energy: {classical_result.total_energies[0].real:.8f} Hartrees")
print(f"Energy Difference: {abs(vqe_result.total_energies[0].real - classical_result.total_energies[0].real):.8f} Hartrees")

# Save optimal parameters
optimal_parameters = vqe_result.raw_result.optimal_parameters
np.save('ch4_frozen_core_optimal_parameters.npy', optimal_parameters)
print("Optimal parameters saved to 'ch4_frozen_core_optimal_parameters.npy'")

# Print out some quantum chemistry information
if hasattr(problem, 'molecule_data'):
    print("\nMolecule data:")
    print(f"Number of molecular orbitals: {problem.molecule_data.num_molecular_orbitals}")
    print(f"Number of electrons: {problem.molecule_data.num_electrons}")
    if hasattr(problem.molecule_data, 'molecular_orbital_energies'):
        print("Molecular orbital energies:")
        print(problem.molecule_data.molecular_orbital_energies)
```
